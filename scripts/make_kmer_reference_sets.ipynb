```python
import pandas as pd
from tqdm import tqdm
```

```python
chromosomes = ["chr1", "chr2", "chr3", "chr4", "chr5", "chr6", "chr7", "chr8", "chr9", "chr10", "chr11", "chr12", "chr16", "chr17", "chr18", "chr19", "chr20"]
```

```python
### load all tagging kmers for each chromosomes, exclude those for cluster specific length and concatenate into one dataframe
tagging_kmer_dfs = []
for chrom in tqdm(chromosomes, total=len(chromosomes), desc="Loading tagging kmers for each chromosome"):
    X = pd.read_csv(f"/home/hain/mnt/korbel/hain/centromere/tagging_kmers/{chrom}_tagging_kmers.tsv", sep="\t")
    tagging_kmer_dfs.append(X[~X["CLUSTER"].str.contains("_HP_")].copy())
all_kmers = pd.concat(tagging_kmer_dfs).reset_index(drop=True)
```

```python
### write all unique kmers to a fasta file for jellyfish querying
with open("/home/hain/mnt/korbel/hain/centromere_genotyping_tool/centromere_genotyping/data/all_tagging_kmers.fasta", "w") as f:
    for kmer in tqdm(all_kmers["KMER"].unique(), total=all_kmers["KMER"].nunique(), desc="Writing unique kmers to fasta file"):
        f.write(f">{kmer}\n{kmer}\n")
```
